```python
a = set ()
for n in range(500, 5001):
    q = bin(n)[2:]
    qcopy = q[1:]
    index = qcopy.find('1')
    if index != -1:
        new = qcopy[index:]
    else:
        new = '0'
    res = n - int(new,2)
    a.add(res)
print(a)
print(len(a))

```

```python
k = []
for n in range(1000, 10000):
    a,b,c,d = map(int, str(n))
    su1 = a+b
    su2 = b+c
    su3 = c+d
    mi, sr, ma = sorted([su1, su2, su3])
    if str(sr) + str(ma) == '511':
        k.append(int(n))
print(min(k), max(k))
```

```python
#24
r = set ()
for n in range(0, 1000):
    s = bin(n)[2:]
    if(s.count('1')%2 == 0):
        s='10' + s[2:] + '0'
    else:
        s='11' + s[2:] + '1'
    if(int(s,2)<35):
        r.add(n)
print(max(r))
```

```python
#29
r = set ()
for n in range(0, 1000):
    s = bin(n)[2:]
    if(n%2 == 0):
        s+='00'
    else:
        s+='11'
    if(int(s,2)>115):
        r.add(n)
print(min(r))
```

```python
#2
k = []
for n in range(700, 801):
    a, b, c = sorted(map(int, str(n)))
    num = []

    for x, y in [(a,b), (a,c), (b,a), (b,c), (c,a), (c,b)]:
        if x != 0:
            num.append(x * 10 + y)


    mi = min(num)
    ma = max(num)
    if ma - mi == 80:
        k.append(int(n))
print(len(k))
```
