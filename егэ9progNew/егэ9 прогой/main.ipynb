```python
f = open("10910.txt")
k=0
for s in f:
    a = list(map(int, s.split()))
    rep=[]
    for x in a:
        if(a.count(x) >= 2):
            rep.append(x)
    if(a.count(min(a)) == 1 and len(rep)>0 and (min(a)+max(a))<sum(rep)):
        k+=1
print(k)



```
